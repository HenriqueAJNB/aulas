# Transformando strings para datetime

datetime -> granularidade ano/mes/dia/hora/minuto/segundo/microsegundo

```python
# Data como string em python
data_as_string = "10/03/2021"
```

```python
# Transformando string para datetime 
data_as_datetime = datetime.strptime(data_as_string, "%d/%m/%Y")
```

Documentação **não oficial** sobre os tipos de formatos: https://www.w3schools.com/python/python_datetime.asp

O formato abaixo deve seguir a string, no caso, dd/mm/aaaa:

%d -> refere-se ao dia com 2 dígitos

%m -> refere-se ao mẽs com 2 dias

%Y -> refere-se ao ano com 4 dígitos

```python
data_as_datetime
# datetime.datetime(2021, 3, 10, 0, 0)
#                  (ano , mes, dia, hora, minuto)
```

# Métodos e atributos do objeto datetime

```python
data_as_datetime.day # Dia 
data_as_datetime.month #Mês
data_as_datetime.year # Ano
data_as_datetime.weekday # Dia da semana (ver documentação para saber qual dia inicial)

data_as_datetime.isocalendar()[1] # Semana do ano (de 1 a 52)
```

```python
datetime.now() # Retorna ano, mês, dia, hora, minuto, segundo e microsegundos atuais
```

# Diferença entre datas (objetos datetime)

```python
data_inicial = datetime(2020, 12, 15)
data_final = datetime(2021, 3, 10)
```

```python
# O resultado gera um objeto timedelta
diff = data_final - data_inicial
diff
```

```python
# Diferença em dias entre as datas
diff.days
```

```python
# Diferença em segundos entre as datas
diff.total_seconds()
```

```python
# Subtraindo 48 dias de uma data
agora = datetime.now()
hoje_menos_45_dias = agora - timedelta(days=48)
hoje_menos_45_dias
```
